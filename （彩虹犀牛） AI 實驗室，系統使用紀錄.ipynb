```python
import matplotlib.pyplot as plt 
from matplotlib.pyplot import figure
import time
import requests_async as requests
import json
import numpy as np
import pandas as pd 
from datetime import datetime

url = "https://rhich.tk/"
#url = "http://0.0.0.0:5000/"
api_route =url+"jupyter/apip/v0"
fetch_user_route = url+"jupyter/fetch_user/v0"

user_payload = {'user_email': user_email, 'user_account_passwords': user_account_passwords}

async def start_api_get(url):
    
    async with requests.Session() as session:
        response = await session.get(url)
        return response
        
async def start_api_post(url, user_payload):

    response = await requests.post(url, data=user_payload)
    return response

async def main(url, post_payload):
    
    #result = await start_api_get(url)
    result = await start_api_post(url, user_payload)
    
    return result

tic = time.perf_counter()
    
data = await main(fetch_user_route, user_payload)

toc = time.perf_counter()


if data.status_code == 200:
    success = data.json()['success']
    if success == True:
        print(data.json()['message']+time.strftime('%X') )
        print("會員編號:", data.json()['會員編號'])
        print("會員email:", data.json()['會員email'])
        print("會員登錄次數:", data.json()['會員登錄次數'])
        print("會員類別:", data.json()['會員類別'])
        print("進階會員訂閱到期日:", data.json()['進階會員訂閱到期日'])
        
        
        
            
        
    
    elif success == False:
        print(data.json()['message']+" "+time.strftime('%X') )
else:
    if data.status_code == 500:
        print("系統無法登錄，請檢查您在步驟一與步驟二的登錄帳號與密碼")
        print("如忘記登入密碼，請至此頁面變更新密碼：https://rhich.tk/reset_password")
        print("如您的帳號為 第一次 登錄此系統，或 第一次 使用本api，請您再按一次乙次上方 ⏩ 按鍵")

    else:
        print(data.status_code)
        print("抱歉，系統有未知的錯誤，請聯繫系統管理員：rainbowrhino88@gmail.com")
        #print(f"data.json()['message'] {time.strftime('%X')}")
        #print(data.json()['message'])
    
```
